### Ejercicio 2:
Escribir un programa que reciba un diccionario con los ingresos
mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.
El resultado final debe ser una serie con los estadísticos.

```python
salario = {} # Creamos dict vacío
r = 'si'

while r != 'no':
    cliente = input('Ingrese nombre del cliente: ') # Ingresar el nombre del cliente
    sueldo = int(input('Ingrese su sueldo: '))
    salario[cliente] = sueldo # agregamos sueldo a cada cliente ingresado
    r = input('Desea continuar?: ') #ingresar si o no

print('\nVisualizamos el diccionario creado:\n' + str(salario))

# Pasamos a Df los valores del dict
df2 = pd.DataFrame(salario.values(),
                  columns = ['Sueldo']
                  )

print('\nVisualizamos estadísticos:')
df2.describe()
```

### Ejercicio 3:
Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

<img src="img1.png">

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
# Generamos Df con números aleatorios

df3 = pd.DataFrame(
    data = [["Enero", 30500, 22000],
            ["Febrero", 35600, 23450],
            ["Marzo", 28300, 18100],
            ["Abril", 33900, 35700],
            ["Mayo", 42500, 32450]], 
    columns = ["Mes", "Ventas", "Gastos"]
    )
df3
```

```python
# 1. Generando nueva columna: Columna1
df3["Columna1"] = (df3['Gastos']/df3['Ventas'])*100
df3
```

```python
# 2. Generando nueva columna: Columna2
df3['Columna2'] = df3['Ventas'].apply(lambda x: 'Meta superada' if x>30000 else 'Meta no superada')
df3
```

```python
# 3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
df3.loc[df3['Gastos'] > df3['Ventas'], 'Columna3'] = df3['Ventas'] - df3['Gastos'] 
df3
```

```python
# 4. Obtener el total de las ventas de los últimos 3 meses
print('Total de ventas de los últimos 3 meses: ' + str(sum(df3['Ventas'].tail(3))))
```

```python
# 5. Obtener el total de gastos de febrero a abril
print('Total de gastos de Feb. a Abr.: ' + str(sum(df3[df3['Mes'].isin(['Febrero','Marzo','Abril'])]['Gastos'])))
```
